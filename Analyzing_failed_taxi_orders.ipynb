```python
#Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
orders = pd.read_csv(filepath_or_buffer="data_orders.csv")
```

```python
orders.head()
```

```python
orders.shape
```

```python
# the random_state argument ensures that we get the same sample
# each time we call the method with the same arguments
orders.sample(n=10, random_state=42)
```

```python
offers = pd.read_csv(filepath_or_buffer="data_offers.csv")
```

```python
offers.shape
```

```python
offers.sample(n=10, random_state=42)
```

```python
df = orders.merge(right=offers, how="inner", on="order_gk")
```

```python
df.sample(n=10, random_state=42)
```

```python
df["is_driver_assigned"] = np.where(df["is_driver_assigned_key"] == 1, "Yes", "No")
df["order_status"] = np.where(df["order_status_key"] == 4, "Client Cancelled", "System Reject")

df.drop(columns=["is_driver_assigned_key", "order_status_key"], inplace=True)
```

```python
df = df.rename(columns={
    "order_datetime": "order_time"})
```

```python
df.sample(n=10, random_state=42)
```

```python
df.groupby(by=["is_driver_assigned", "order_status"])["order_gk"].count()
```

```python
df_q1 = df.pivot_table(columns=["is_driver_assigned", "order_status"], values="order_gk", aggfunc="count")

# Plot using Matplotlib
fig, ax = plt.subplots(figsize=(7, 7))
df_q1.plot(kind="bar", ax=ax, legend=True, rot=0)
plt.ylabel("Order Count")
plt.title("Orders by Driver Assignment and Status")
plt.show()
```

```python
# the structure of the pivot table is very similar to the group-by dataframe
df_q1
```

```python
# extract hour from the time column
df["order_hour"] = df["order_time"].str.split(":").apply(lambda split: split[0])
```

```python
# print a small sample to make sure that the transformation is correct
df.sample(n=5, random_state=42)
```

```python
order_counts = df.groupby(by="order_hour")["order_gk"].count()

fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(order_counts.index, order_counts.values, marker="o", linestyle="-", label="Failed Orders")

ax.set_xticks(range(0, 24))
ax.set_xlabel("Hour of the Day")
ax.set_ylabel("Order Count")
ax.set_title("Count of Failed Orders by Hour of Day")
ax.legend()

plt.grid(True)
plt.show()
```

```python
grouped_q2 = df.groupby(by=["order_hour", "is_driver_assigned", "order_status"])["order_gk"].count()
grouped_q2
```

```python
df_q2 = grouped_q2.reset_index().pivot(index="order_hour",
                                       columns=["is_driver_assigned", "order_status"],
                                       values="order_gk")

fig, ax = plt.subplots(figsize=(13, 7))
for col in df_q2.columns:
    ax.plot(df_q2.index, df_q2[col], marker="o", linestyle="-", label=f"{col}")

ax.set_xticks(range(0, 24))
ax.set_xlabel("Hour of the Day")
ax.set_ylabel("Order Count")
ax.set_title("Count of Failed Orders Per Hour and Category")
ax.legend(title="Driver Assigned & Order Status", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)

plt.show()
```

```python
grouped_q3 = df.groupby(by=["order_hour", "is_driver_assigned"])["cancellations_time_in_seconds"].mean()
grouped_q3
```

```python
df_q3 = grouped_q3.reset_index().pivot(index="order_hour",
                                       columns="is_driver_assigned",
                                       values="cancellations_time_in_seconds")
fig, ax = plt.subplots(figsize=(13, 7))

for col in df_q3.columns:
    ax.plot(df_q3.index, df_q3[col], marker="o", linestyle="-", label=f"Driver Assigned: {col}")

ax.set_xticks(range(0, 24))
ax.set_xlabel("Hour of the Day")
ax.set_ylabel("Avg. Time to Cancellation (Seconds)")
ax.set_title("Average Time to Cancellation Per Hour and Driver Assignment")
ax.legend(title="Driver Assigned", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)

plt.show()
```

```python
avg_eta_per_hour = df.groupby(by="order_hour")["m_order_eta"].mean()

fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(avg_eta_per_hour.index, avg_eta_per_hour.values, marker="o", linestyle="-", label="Avg. ETA")

ax.set_xticks(range(0, 24))
ax.set_xlabel("Hour of the Day")
ax.set_ylabel("Average ETA (Seconds)")
ax.set_title("Average ETA per Hour")
ax.legend()
plt.grid(True)

plt.show()
```
